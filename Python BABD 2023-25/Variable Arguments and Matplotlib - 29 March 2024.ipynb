## Matplotlib

* Documentation at: https://matplotlib.org/stable/index.html

### T1 - Introduction

#### Basic Overview
* Visualization library in Python
* Introduced by John Hunter in 2002
* Designed on top of NumPy arrays
* Designed to work with the broader SciPy stack
* Comprises several plots like line, bar, scatter, histogram, etc. as well as display of images   
* Use import matplotlib.pyplot as py to import it as a library
* One of the most popular cross-platform library for making 2D plots from data in arrays
    - Make use of plot() for plotting; and show the plot using the show() 

#### Installation
* Installed by default during Anaconda setup
* Using python package manager, pip
    - pip install matplotlib
* Using Anaconda prompt, if needed
    - conda install matplotlib
   
 
#### Pyplot in Matplotlib

* Is a module in Matplotlib, which provides a MATLAB like interface with the additional benefit of using Python as a base and being free and open-source.
* Comprises a rich collection of functions used for making changes to the figure like creating figure, creating plot area in figure, plotting line in plot area, decorating plot with labels etc.
* Supports various plots like Line Plot, Histogram, Scatter, 3D Plot, Image, Contour, and Polar.
* Syntax:
    matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
* Parameters: This function accepts parameters that enables us to set axes scales and format the graphs. These parameters are mentioned below :-
    - plot(x, y): plot x and y using default line style and color.
    - plot.axis([xmin, xmax, ymin, ymax]): scales the x-axis and y-axis from minimum to maximum values
    - plot.(x, y, color=’green’, marker=’o’, linestyle=’dashed’, linewidth=2, markersize=12): x and y co-ordinates are marked using circular markers of size 12 and green color line with — style of width 2
    - plot.xlabel(‘X-axis’): names x-axis
    - plot.ylabel(‘Y-axis’): names y-axis
    - plot(x, y, label = 'Sample line') plotted Sample Line will be displayed as a legend

### Matplotlib MARKERS 
    - use marker keyword to specify a specific marker type during plotting; ike 'o' for circle; marker='*'
    = fmt comprising marker type:linestyle:color can be used during plot; 'o:r'
    - markersize or ms can be used to specify size of marker; ms=20
    - use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers
    - use the keyword argument markerfacecolor or the shorter mfc to set the color of the face/core of the markers
