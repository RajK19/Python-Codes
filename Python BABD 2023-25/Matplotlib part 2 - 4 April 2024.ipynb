#### Pyplot in Matplotlib

* Is a module in Matplotlib, which provides a MATLAB like interface with the additional benefit of using Python as a base and being free and open-source.
* Comprises a rich collection of functions used for making changes to the figure like creating figure, creating plot area in figure, plotting line in plot area, decorating plot with labels etc.
* Supports various plots like Line Plot, Histogram, Scatter, 3D Plot, Image, Contour, and Polar.
* Syntax:
    matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
* Parameters: This function accepts parameters that enables us to set axes scales and format the graphs. These parameters are mentioned below :-
    - plot(x, y): plot x and y using default line style and color.
    - plot.axis([xmin, xmax, ymin, ymax]): scales the x-axis and y-axis from minimum to maximum values
    - plot.(x, y, color=’green’, marker=’o’, linestyle=’dashed’, linewidth=2, markersize=12): x and y co-ordinates are marked using circular markers of size 12 and green color line with — style of width 2
    - plot.xlabel(‘X-axis’): names x-axis
    - plot.ylabel(‘Y-axis’): names y-axis
    - plot(x, y, label = ‘Sample line ‘) plotted Sample Line will be displayed as a legend
    
#### What is Plotting?
 - The plot() function is used to draw points (markers) in a diagram.
 - The function takes parameters for specifying points in the diagram.
    - Parameter 1 is an array containing the points on the x-axis.
    - Parameter 2 is an array containing the points on the y-axis.
 - To plot a line from (1, 9) to (7, 15), we have to pass two arrays [1, 7] and [9, 15] to the plot function.
    

```python
import matplotlib

print(matplotlib.__version__)


# import the lib
import matplotlib.pyplot as plt
plt.style.use('classic')

#Available styles in matplotlib
print(plt.style.available)

# Creating a line plot

x_values=[2,3,5,6,7,9] # x axis values
y_values=[4,9,24,35,48,78] # y axis values
plt.plot(x_values, y_values)
plt.show()  
```

```python
#In terminal mode or script, show() is needed to display
#whereas in interactive mode, it is irrelevant

#Drawing complete line plot
# complete line plot
#To add the x axis label...plt.xlabel('')
#To add the y axis label...plt.ylabel('')
#To add the Title...plt.title('')

x_values=list(range(2,10,2)) # x axis values
y_values=[i**2 for i in x_values] # y axis values
plt.xlabel('Speed (miles/hour)')
plt.ylabel('Distance (miles)')
plt.title('Speed versus Distance Plot')
plt.plot(x_values, y_values, 'b', linewidth=2, marker='o')
plt.show()
#plt.savefig('babd1.jpeg')  #to save plot
```

```python
import matplotlib.pyplot as plt
data={'apple':10, 'orange':15, 'Mango':20, 'lemon':7}

x=list(data.keys())
y=list(data.values())

fig,axs=plt.subplots(1,3, figsize=(12,3), sharey=True)

axs[0].bar(x, y)
axs[0].tick_params(axis='x', labelrotation = 30)

axs[1].scatter(x, y)
axs[1].tick_params(axis='x', labelrotation = 30)

axs[2].plot(x, y)
axs[2].tick_params(axis='x', labelrotation = 30)

fig.suptitle('Categorical Plotting')
```

```python
import matplotlib.pyplot as plt
data={'apple':10, 'orange':15, 'Mango':20, 'lemon':7}
plt.style.use('classic')
x=list(data.keys())
y=list(data.values())

fig,axs=plt.subplots(1,3, figsize=(12,3), sharey=True)

axs[0].bar(x, y)
axs[0].tick_params(axis='x', labelrotation = 30)

axs[1].scatter(x, y)
axs[1].tick_params(axis='x', labelrotation = -40)

axs[2].plot(x, y)
axs[2].tick_params(axis='x', labelrotation = 30)

fig.suptitle('Categorical Plotting')
```
