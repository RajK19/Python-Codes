### Matplotlib LABELS and TITLE 

    - Use the xlabel() and ylabel() functions in pyplot to set a label for the x- and y-axis.
    - use the title() function to set a title for the plot.
    - use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels
    - use the loc parameter in title() to position the title. Legal values are: 'left', 'right', and 'center'. Default value is 'center'.
    

```python
x = np.array([80, 82, 84, 86, 88, 90, 92, 94, 96, 98])
y = np.array([340, 350, 360, 370, 380, 390, 400, 410, 420, 430])

plt.plot(x, y, "D-.b")

#Adding labels to x and y axis
plt.xlabel("Average Heartbeat")
plt.ylabel("Calories Consumed")


#Setting up title for the plot
plt.title("Marathon Performance Statistics")
#plt.title("Marathon Performance Statistics", loc = 'right')

plt.show()
```

### Matplotlib - Scatter
- use the scatter() function to draw a scatter plot.
    * scatter() function plots one dot for each observation(x,y). 
    * Involves two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis
- Multiple scatter plots are plotted with different colors, by default blue and orange and so on. Comparison of Scatter Plots can be done by plotting different sets of data on same plot.
- color or 'c' argument can be used to set up the color for each scatter plot
- Only 'c' argument can be used (not color) and assigned array of colors, if colors need to assigned at dot level for different dots in a scatter plot.
- ColorMap can be assigned to different scatter plots using cmap argument.
    * Matplotlib module has a number of available colormaps. A colormap is like a list of colors, where each color has a value that ranges from 0 to 100. For example, one of the colormaps known as 'viridis' ranges from 0(purple color) to 100 (yellow color).
    * Include colormap using plt.colorbar() statement.
- Change the size of the dots with the s argument. We need to make sure that just like colors, array for sizes has the same length as the arrays for the x- and y-axis.
- Use alpha argument to adjust the transparency of the dots. 

```python
#Matplotlib - Scatter plots

import matplotlib.pyplot as plt
import numpy as np

#1. Simple scatter plot - Age versus speed of car; As age increases (x-axis); speed decreases (y-axis)

#Day 1 data
x = np.array([5,7,9,11,8,6])   #Age of car
y = np.array([99,95,91,86,94,98])  #Speed of car
plt.scatter(x, y)

#Day 2 data
x = np.array([3,5,7,9,6,4])   #Age of car
y = np.array([120,110,100,90,95,115])  #Speed of car
plt.scatter(x, y)

plt.show()
```

### Matplotlib Bars
- Use the bar() function to draw bar graphs with pyplot
    - takes arguments that describes the layout of the bars.
    - The categories and their values represented by 1st and 2nd argument as arrays.
- Use barh() for displaying horizontal bars
- Use the following keyword argument in bar() or barh():
    * color for setting color of the bars
    * height and width for setting height and width of the bars respectively
        - height is used in the case of horizontal bars, barh()
        - width is used for vertical bars created using bar(). Default width is 0.8

```python
#Bar Charts

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array(["L1", "L2", "L3", "L4"])
y1 = np.array([-13, 18, -11, 20])

plt.bar(x1,y1, color='green', width=0.4)  #hex value can be used as well - '#4CAF50'
plt.show()

plt.barh(x1,y1, color='yellow', height=0.2)
plt.show()
```
