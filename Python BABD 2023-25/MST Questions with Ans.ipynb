```python
import numpy as np

arr1 = np.array([7, 5, 3, 2, 6, 1, 4])

#Sort the numpy array in ascending order
arr2 = np.sort(arr1)

#Sort the numpy array in descending order
arr3 = arr2[::-1]

print("arr1: ", arr1)

print("arr2: ", arr2)

print("arr3: ", arr3)
```
